```python
# Step 1: Install and Import Libraries
!pip install statsmodels --quiet
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from statsmodels.tsa.seasonal import seasonal_decompose
```

```python
# Step 2: Load Real-World Dataset
# We'll use the famous Airline Passengers dataset (Monthly totals of international airline passengers, 1949–1960)
url = "https://raw.githubusercontent.com/jbrownlee/Datasets/master/airline-passengers.csv"
df = pd.read_csv(url, parse_dates=['Month'], index_col='Month')
```

```python
# Display first few rows
print("Sample Data:")
print(df.head())
```

```python
# Step 3: Plot the Original Time Series
plt.figure(figsize=(12,5))
plt.plot(df, label="Airline Passengers")
plt.title("Monthly Airline Passengers (1949–1960)")
plt.xlabel("Date")
plt.ylabel("Number of Passengers")
plt.legend()
plt.show()
```

```python
# Step 4: Apply Additive Decomposition
additive_decomposition = seasonal_decompose(df['Passengers'], model='additive', period=12)


plt.figure(figsize=(12,8))
additive_decomposition.plot()
plt.suptitle("Additive Decomposition", fontsize=16)
plt.show()
```

```python
# Step 5: Apply Multiplicative Decomposition
multiplicative_decomposition = seasonal_decompose(df['Passengers'], model='multiplicative', period=12)

plt.figure(figsize=(12,8))
multiplicative_decomposition.plot()
plt.suptitle("Multiplicative Decomposition", fontsize=16)
plt.show()
```
